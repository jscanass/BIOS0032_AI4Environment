## Lists

A **list** in Python is an **ordered** and **mutable** collection of values.
‚ÄúOrdered‚Äù means that elements have a fixed position, and ‚Äúmutable‚Äù means that the list can be changed after it is created.

```python
my_list = ["b", "a", "d", "f", "e", "c"]

print(len(my_list))  # length (number of elements)

# accessing elements of a list
print(my_list[0])  # index starts at zero!
print(my_list[0:3])  # from...to (excluding last)
print(my_list[-1])  # last element
print(my_list[1:4:2])  # from..to, every 2nd
```

üí° **Important:** Indices in Python start at zero.
This is a very common source of confusion, don't forget it!

Since lists are mutable, we can easily add, remove, reorder, and combine elements:

```python
my_list.append("g")  # append element at the end
print(my_list)

my_list.insert(1, "ab")  # insert element at index (again: they start at zero)
print(my_list)

my_list.remove("b")  # remove element
print(my_list)

my_list.sort()  # sort in ascending order
print(my_list)
```

Built-in functions:

```python
print(len(my_list))  # length of list, str, etc.; anything with more than one value

print(sum([1, 2, 3]))
print(max([4, 5, 6]))
```

## If statements

Conditional statements (`if`) let you execute different code depending on conditions:

```python
# Code inside if statements will only execute if the condition is true
if "Apple" in my_dict:
    print("I don't like apple")

# You can add an `else` block, that will run if the condition is not true
if len(my_list) > 20:
    print("That's a lot of things")
else:
    print("List under control")

# You can also add `elif` (short for else if) blocks to check against additional conditions
if "Daikon" in my_dict:
    print("Cabbage salad tonight!")
elif len(my_dict) > 3:
    print("More than three items in the pantry.")
else:
    my_dict = {}
    print("Just threw out everything from the kitchen.")
```

This means:

- All lines belonging to the same block must have the **same level of indentation**.

## Functions

Functions are reusable blocks of code.
They let you **encapsulate a common workflow** and give it a name, which makes your code easier to organise and understand.

If you find yourself repeating the same code multiple times, a function allows you to **define it once** and reuse it whenever needed.

In Python, you define a function using the keyword `def`:

```python
# Give the function a descriptive name that explains what it does
def select_positives(number_list):
    # `number_list` is the input to the function
    # It should be a list of numbers

    # Inside the function, you can perform any operations on the input

    selected_numbers = []
    for number in number_list:
        if number >= 0:
            selected_numbers.append(number)

    # Return the result to the caller
    return selected_numbers
```

You can **call the function** like this:

```python
my_list = [-20, 10, -15, 5, 1]

print("Original list:")
print(my_list)

# call function, store return argument into a new variable
my_list_filtered = select_positives(my_list)

print("Filtered list:")
print(my_list_filtered)
```

You can also:
- Specify multiple input arguments
- Give default values to them
- Return multiple outputs

for example:

```python
def split_train_test(
    data,
    ratio_train=0.6,
    ratio_val=0.3,
):
    """
    Receives an input list and splits it into training, validation, and test sets
    """
    num_total = len(data)
    num_train = int(ratio_train * num_total)
    num_val = int(ratio_val * num_total)

    # Slice the list to create the subsets
    # list[start:end] selects elements from index 'start' up to (not including) 'end'
    data_train = data[0:num_train]
    data_val = data[num_train : num_train + num_val]
    data_test = data[num_train + num_val :]  # slice from after validation to the end

    return data_train, data_val, data_test


# to call the function and store all of its return values
d_train, d_val, d_test = split_train_test(my_list, ratio_train=70)
```

Notes:

- Arguments with default values can be omitted.
- You can specify inputs explicitly by name (`ratio_train=70`) or positionally (in the order they appear in the function definition).
- The function returns three outputs, which can be assigned to three variables.

## Handling errors

Sometimes, something can go wrong when your code runs (for example, dividing by zero or accessing an element that doesn't exist in a list).
Python will normally **throw an error**, which stops the program.

You can handle these errors gracefully using **try-except statements**.

A `try-except` statement:

1. **Tries** to run a block of code.
2. If any line in that block causes an **exception** (an error), Python jumps to the `except` block instead of crashing.

```python
try:
    # will try to execute this code block, line-by-line
    index = 0
    print(my_list[index])

except Exception as e:
    # if and only if any line above doesn't work (it raises an exception), this block is executed
    print("An error occurred:")
    print(e)
```

### Accessing & editing dataframe values

In a pandas DataFrame, each row and each column has an associated **index**.
You can use these indices to select, inspect, or modify individual values or entire subsets of the data.

- Columns are usually indexed by their names.
- Rows are usually indexed by their row number (starting at 0), unless a custom index is used.

```python
# Get a specific entry in the dataframe

# Get value by indices
value = iris.loc[10, "species"]  # value @ row with index 10, species column

# Get value by numerical position
value2 = iris.iloc[10, 4]  # value @ 10th row, 5th column (Python is 0-indexed)

# A combination of the previous approaches
value3 = iris["species"][10]  # 10th value in the species column

# All are equivalent ways of retrieving the same value
print(value == value2 == value3)
```

```python
# Edit a single entry
iris.loc[0, "sepal_length"] = 10

# Create a new column
iris["petal_area"] = iris["petal_width"] * iris["petal_length"]
```

### Compute on every row

Sometimes you want to compute a new value based on multiple columns in each row.
One way to do this is to define a function that operates on a single row.

```python
def compute_total_area(row):
    sepal_area = row["sepal_length"] * row["sepal_width"]
    petal_area = row["petal_length"] * row["petal_width"]
    return sepal_area + petal_area
```

A very explicit (but slower) way to apply this function is to iterate over rows:

```python
# Iterate over every row
total_area = 0
for index, row in iris.iterrows():
    # Sum all areas
    total_area += compute_total_area(row)
```

A more efficient approach in pandas is to use `.apply()`:

```python
# Apply a function to each row, get a Series of results
iris_areas = iris.apply(compute_total_area, axis=1)

# See results
iris_areas.head()
```

### Grouping by values

Another powerful feature of pandas is grouping.
Grouping allows you to split a DataFrame into subsets based on the values of one or more columns, perform computations within each group, and then combine the results.

In the example below, we group the rows by the species column and compute the standard deviation of each numerical column within each species:

```python
# Group dataframe rows by species and compute the standard deviation of column values within each group.
iris.groupby("species").std()
```

You can also apply custom computations to each group.
For example, we can compute the total area (using the function defined earlier) and sum it separately for each species:

```python
# Group dataframe by species and compute the mean total area for each group
iris.groupby("species").apply(
    lambda group: group.apply(compute_total_area, axis=1).sum()
)
```

üí° **Note on lambda functions**

A lambda function is a small, one-line function without a name.
It is useful for short operations that you only need once.

`lambda x: x ** 2`

In the example above, the `lambda` function is applied once per group created by `groupby()`.
If the logic becomes long or hard to read, it's better to define a regular function using `def`.

### Creating a simple scatter and line plot

```python
import matplotlib.pyplot as plt  # This is the common way to import and abbreviate matplotlib
```

```python
# Create an independent variable, ranging from 0 to 10
x = np.linspace(0, 10, 100)

# Create a dependent variable
y = np.cos(x)

# Add some normal noise
z = y + np.random.normal(scale=0.1, size=100)
```

```python
# Create a new figure
plt.figure(figsize=(6, 3))  # Adjust the size

# Create a line plot
plt.plot(x, y)
```

```python
# Create another figure
plt.figure(figsize=(6, 3))  # Adjust the size

# Create a scatter plot
plt.scatter(x, z)
```

```python
plt.figure(figsize=(6, 3))

# Create a red line plot.
plt.plot(x, y, label="line plot", color="red")

# Overlay the scatter plot with crosses as markers
plt.scatter(x, z, label="scatter plot", marker="x")

# Add legend to figure
plt.legend()
```

### Modifying titles and axes labels

```python
plt.figure(figsize=(6, 3))

# Create a red line plot.
plt.plot(x, y, label="prediction", color="red")

# Overlay the scatter plot with crosses as markers
plt.scatter(x, z, label="measurements", marker="x")

# Add legend to figure
plt.legend()

# Modify the label of the x-axis
plt.xlabel("Time (days)")

# Modify the label of the y-axis
plt.ylabel("Displacement (m)")

# Add title to the figure
plt.title("Displacement time series")
```

### Creating compound figure

```python
plt.figure(figsize=(12, 3))

# Create a 1 row, 2 column, grid of subplots and select the first subplot.
ax1 = plt.subplot(1, 2, 1)

# Create line plot in subplot
ax1.plot(x, y, label="prediction", color="red")

# Overlay the scatter plot with crosses as markers
ax1.scatter(x, z, label="measurements", marker="x")

# Add legend to figure
ax1.legend()

# Modify the label of the x-axis. Note a difference?
ax1.set_xlabel("Time (days)")

# Modify the label of the y-axis
ax1.set_ylabel("Displacement (m)")

# Add title to the figure
ax1.set_title("Displacement time series")

# Add subplot. This is the second subplot in the grid of subplots.
ax2 = plt.subplot(1, 2, 2)

# Compute the prediction error
error = z - y

# Plot a histogram of the error in the second subplot
ax2.hist(error, label="Prediction error")

# Add labels to axis
ax2.set_xlabel("Error (m)")

# Set subplot title
ax2.set_title("Histogram of prediction errors")
```

### Adding text

```python
plt.figure(figsize=(6, 3))

# Select a single point in the data
point = x[31], z[31]

# Create a red line plot.
plt.plot(x, y, label="line plot", color="red")

# Overlay the scatter plot with crosses as markers
plt.scatter(x, z, label="scatter plot", marker="x")

# Show selected point as a large black dot
plt.scatter(
    [point[0]],
    [point[1]],
    label="test point",
    marker="o",
    s=100,
    color="black",
)

# Add formula for the prediction. Anchor the text at the point x= 5.5, y=0.
plt.text(5.5, 0, "y = cos(x)", color="red")

# Add text and arrow pointing to the test point
plt.annotate(
    "test point",
    point,
    xytext=[point[0], 0.5],
    color="orange",
    arrowprops={"color": "orange", "width": 1},
)

# Add legend to figure
plt.legend()
```
